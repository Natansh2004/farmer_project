```python
# Data transformation
# scaling down the data (in a range)
# to remove all the outliers 

from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
x_transformed = std_scaler.fit_transform(x)  
# fit_transform()--> fit as well as tranform the data
x_transformed

# std_scaler uses z_score formula behind the scenes
```

```python
# we've scaled down the data but still both the graphs are exactly same
# distribution of both the graphs are similar
# just we have scaled down the data into a another form
# outliers will be hidden
# the meaning of the data is retained

# k_means algorithm is such an algorithm that demands outliers-free data

# if we do not scale down the data before passing this to 
# algorithm, then the accuracy will reduce

# and we will use this algorithm in this project
```

```python
## NOW THE DATA IS READY TO TRAIN THE ALGORITHM

from sklearn.cluster import KMeans
# KMeans algorithm performs clustering analysis
```

```python
kmeans_obj = KMeans()
kmeans_obj
```

```python
labels = kmeans_obj.fit_predict(x_transformed)

# fit_predict ---> fit and predict both
# algorithm trained and predict
```
